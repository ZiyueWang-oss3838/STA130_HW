```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import statsmodels.formula.api as smf

# Load the dataset
url = "https://raw.githubusercontent.com/KeithGalli/pandas/master/pokemon_data.csv"
pokeaman = pd.read_csv(url)

# Replace NaN values in the "Type 2" column with "None"
pokeaman.fillna('None', inplace=True)

# Split the data into training and testing sets
np.random.seed(130)
fifty_fifty_split_size = int(pokeaman.shape[0] * 0.5)
pokeaman_train, pokeaman_test = train_test_split(pokeaman, train_size=fifty_fifty_split_size)

# Define the model
model5_linear_form = 'HP ~ Attack + Defense + Speed + Legendary + Q("Sp. Def") + Q("Sp. Atk") + C(Generation) + C(Q("Type 1")) + C(Q("Type 2"))'
model5_spec = smf.ols(formula=model5_linear_form, data=pokeaman_train)
model5_fit = model5_spec.fit()

# Display the summary
print(model5_fit.summary())

yhat_model5 = model5_fit.predict(pokeaman_test)
y = pokeaman_test.HP
print("'In sample' R-squared:    ", model5_fit.rsquared)
print("'Out of sample' R-squared:", np.corrcoef(y,yhat_model5)[0,1]**2)
```

Model 5:
Rationale: In model5_linear_form, the main prefictors are Speed, Sp. Def, Attack, Defense, and Sp. Atk, categorical indicators are Type 1, Type 2, Legendary, and Generation.
Objective: The model5_linear_form use a lot of prefictors to capture the effect on HP, then it establish initial comprehensive baseline for prediction.

```python
# Here's something a little more reasonable...
model6_linear_form = 'HP ~ Attack + Speed + Q("Sp. Def") + Q("Sp. Atk")'
# And here we'll add the significant indicators from the previous model
# https://chatgpt.com/share/81ab88df-4f07-49f9-a44a-de0cfd89c67c
model6_linear_form += ' + I(Q("Type 1")=="Normal")'
model6_linear_form += ' + I(Q("Type 1")=="Water")'
model6_linear_form += ' + I(Generation==2)'
model6_linear_form += ' + I(Generation==5)'

model6_spec = smf.ols(formula=model6_linear_form, data=pokeaman_train)
model6_fit = model6_spec.fit()
model6_fit.summary()

yhat_model6 = model6_fit.predict(pokeaman_test)
y = pokeaman_test.HP
print("'In sample' R-squared:    ", model6_fit.rsquared)
print("'Out of sample' R-squared:", np.corrcoef(y,yhat_model6)[0,1]**2)
```

Model 6:
Developement: model6_linear_form focuses on the most impactful prefictors which are: Speed, Attack, Sp. Def, Sp. Atk, and includes the significant categorical indictors which are: Generations (2 and 5) and Type 1 (Normal and Water), etc.
Rationale:
The model refines predictor through eliminates less significant variables in order to reduce complexity. This improves model interpretability and prediction accuracy.

```python
# And here's a slight change that seems to perhaps improve prediction...
model7_linear_form = 'HP ~ Attack * Speed * Q("Sp. Def") * Q("Sp. Atk")'
model7_linear_form += ' + I(Q("Type 1")=="Normal")'
model7_linear_form += ' + I(Q("Type 1")=="Water")'
model7_linear_form += ' + I(Generation==2)'
model7_linear_form += ' + I(Generation==5)'

model7_spec = smf.ols(formula=model7_linear_form, data=pokeaman_train)
model7_fit = model7_spec.fit()
model7_fit.summary()
yhat_model7 = model7_fit.predict(pokeaman_test)
y = pokeaman_test.HP
print("'In sample' R-squared:    ", model7_fit.rsquared)
print("'Out of sample' R-squared:", np.corrcoef(y,yhat_model7)[0,1]**2)
```

```python
# And here's a slight change that seems to perhas improve prediction...
model7_linear_form_CS = 'HP ~ scale(center(Attack)) * scale(center(Speed))'
model7_linear_form_CS += ' * scale(center(Q("Sp. Def"))) * scale(center(Q("Sp. Atk")))'
# We DO NOT center and scale indicator variables
model7_linear_form_CS += ' + I(Q("Type 1")=="Normal")'
model7_linear_form_CS += ' + I(Q("Type 1")=="Water")'
model7_linear_form_CS += ' + I(Generation==2)'
model7_linear_form_CS += ' + I(Generation==5)'

model7_CS_spec = smf.ols(formula=model7_linear_form_CS, data=pokeaman_train)
model7_CS_fit = model7_CS_spec.fit()
model7_CS_fit.summary().tables[-1] 
# "Cond. No." is NOW 15.4 due to centering and scaling

# "Cond. No." WAS 2,340,000,000 WITHOUT to centering and scaling
model7_fit.summary().tables[-1]
```

Model 7:
Further Refinement and Interaction Terms: model7_linear_form introduces interaction terms between Attack, Speed, Sp. Def, and Sp. Atk alongside significant indicators.
Objective: Adding interactions allows the model to capture potential synergistic effects between continuous predictors, and this explains more variability in HP.

Model 7 with centering and scaling:
Multicollinearity Addressed: model7_linear_form_CS introduces interaction terms between Speed, Attack, Sp. Def, and Sp. Atk alongside significant indicators.
Outcome: Scaling and centering to continuous predictors improves robustness by mitigating the effects of multicollinearity. This enhances out-of-sample performance stability.

summary and links:

In this progression of linear models, each version builds upon the last to improve predictive accuracy and model stability:

Model 5 includes a wide range of variables, providing a comprehensive baseline with various attributes (e.g., Attack, Defense, Speed) and categorical factors (Generation, Type 1, Type 2).

Model 6 refines Model 5 by excluding certain variables and adding indicator terms for significant types (e.g., "Normal", "Water") and specific generations (Generation == 2 and Generation == 5), allowing a more targeted approach to capture relevant predictors without unnecessary complexity.

Model 7 introduces multiplicative interactions between key predictors (e.g., Attack * Speed * Sp. Def * Sp. Atk) to account for nonlinear relationships, enhancing model flexibility. A variant with centered and scaled terms (Model 7_CS) reduces multicollinearity, resulting in a lower condition number and improved stability.

This progression results in a streamlined model that balances complexity, predictive strength, and stability by refining variable selection, interactions, and scaling techniques.
